### Creating new calculated fields in our data frame
#### 1.  Calculating the total number of children 'totChildren' by adding 'noOfKids' and 'noOfTeens'

```python
data['totChildrens'] = data['noOfKids'] + data['noOfTeens']
```

#### 2. Extracting year from 'dtEnrolled' to identify the Year the customer started using the service

```python
data['yrEnrolled'] = pd.DatetimeIndex(data['dtEnrolled']).year
```

#### 3. Calculating the total spending 'totSpendings' by summing all columns containing keyword 'amt'

```python
amtCols = [col for col in data.columns if 'amt' in col]
amtCols
```

```python
data['totSpendings'] = data[amtCols].sum(axis=1)
```

#### 4. Calculating the total purchses made 'totPurchases' by summing all columns containing keyword 'Purchases'

```python
purchasesCols = [col for col in data.columns if  'Purchases' in col]
data['totPurchases'] = data[purchasesCols].sum(axis=1)
```

#### 5. Finding total numbers of campaigns accepted for each customers by adding columns containing 'campaign' & 'offerAccepted'

```python
campaignsCols = [col for col in data.columns if 'campaign' in col] + ['offerAccepted']
data['totAcptCampaigns'] = data[campaignsCols].sum(axis=1)
```
